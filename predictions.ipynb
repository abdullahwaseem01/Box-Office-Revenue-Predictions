# Movie Revenue Predictions using Linear Regression 

### Import libraries and data

```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
import numpy as np

data = pd.read_csv('data\cost_revenue_clean.csv')
```

```python
data
```

```python
data.describe()
```

### Create dataframes from imported data

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

### Plot budget vs. revenue on scatter plot 

```python
plt.figure(figsize=(25,5))
plt.scatter(X,y, alpha=(0.5))
plt.title('Budget vs. Revenue')
plt.xlabel('Production Budget($)')
plt.ylabel('Global Revenue($)')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

### Initialize regression model and fit with budget and revenue data

```python
regression = LinearRegression()
regression.fit(X, y)
```

### Regression line slope

```python
'\u03F41: ' +  str(regression.coef_[0,0])
```

### Regression line intercept

```python
'\u03F40: ' + str(regression.intercept_[0])
```

### Plotting the regression model using budgets to predict revenue values

```python
plt.figure(figsize=(25,5))
plt.scatter(X,y, alpha=(0.5))
plt.plot(np.asarray(X), np.asarray(regression.predict(X)), color='red', linewidth=3)
plt.title('Budget vs. Revenue')
plt.xlabel('Production Budget($)')
plt.ylabel('Global Revenue($)')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

### Testing the accuracy of the regression model

```python
"The regression model is " + str(round(regression.score(X, y) * 100, 0)) + "%" + " accurate in performing the predictions for movie revenue."
```
